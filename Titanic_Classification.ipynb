```python
from pycaret.classification import *

data = train_yes.drop('Survived',axis=1)
target = train_yes['Survived']

setup(data=data,target=target)

compare_models()
```

```python
from sklearn.model_selection import train_test_split

X = train_yes.drop('Survived',axis=1)
y = train_yes['Survived']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc = GradientBoostingClassifier()

gbc.fit(X_train, y_train)
gbc_pred = gbc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(classification_report(y_test,gbc_pred))
```

```python
print(confusion_matrix(y_test,gbc_pred))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
gbc = GradientBoostingClassifier()

# Define the parameter grid for GridSearchCV
param_grid = {
    'n_estimators': [50, 100, 150],
    'learning_rate': [0.001, 0.05, 0.1, 0.5, 1],
    'max_depth': [3, 6, 9],
    'min_samples_split': [2, 3, 4, 5],
    'min_samples_leaf': [1, 2, 3],
    'loss': ['log_loss', 'deviance', 'exponential'],
    'subsample': [0.5, 1.0]
}
```

```python
gscv = GridSearchCV(estimator=gbc, param_grid=param_grid, cv=5, n_jobs=-1)
```

```python
gscv.fit(X_train,y_train)
```

```python
params = gscv.best_params_
```

```python
gbr1 = GradientBoostingClassifier(**params)
```
